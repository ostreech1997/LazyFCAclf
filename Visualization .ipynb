```python
from LazyFCA import preprocessing, Predict, Metrics
```

```python
import matplotlib.pyplot as plt
```

### Preprocessing data step

```python
data_dict = preprocessing()
```

# Quality evaluation of model depending on parameter max_int

```python
max_int = [1,2,3,4,5]
accuracy = []
precision = []
recall = []
ROC_AUC = []
for i in max_int:
    Y_pred = Predict(data_dict, i)
    metrics = Metrics(data_dict['Y_test'],Y_pred)
    accuracy.append(metrics['accuracy'])
    precision.append(metrics['precision'])
    recall.append(metrics['recall'])
    ROC_AUC.append(metrics['ROC_AUC'])
```

```python
plt.figure(figsize=(12,7))
plt.plot(max_int, accuracy, color='blue', label='accuracy')
plt.plot(max_int, precision, color='green', label='precision')
plt.plot(max_int, recall, color='black', label='recall')
plt.plot(max_int, ROC_AUC, color='red', label='ROC AUC')
plt.legend()
plt.xlabel('max_int')
plt.ylabel('quality')
plt.title('Some metrics to estimate model quality')
plt.show()
```
